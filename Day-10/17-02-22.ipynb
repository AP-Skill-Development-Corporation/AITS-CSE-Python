### Object
- An object is also called as an instance of a class and the process of creating this object is called instantiation
- syntax: obj_name=class_name

````python
# Example for object creation
class Hi:
    a=10
    b=20
    def display():
        print('Hi i am freom display method')
obj=Hi  # object creation
print(obj.a)
print(obj.b)
obj.display()
````

````python
# Example2 for object creation
class Math():
    '''inside this method i am going to defining add() and mul()'''
    def add(num1,num2):
        return num1+num2
    def mul(num1,num2):
        return num1*num2
obj=Math
print(obj.add(12,13))
print(obj.mul(2,4))
````

### Inheritance in python
- a class that inherits all the method and properties from another class
- parent class is also called as base class
- child class is the class that inherits from another class so child class is also called as derived class
- syntax:  class ChildClass(parentClass):

## Single inheritance

````python
class ClassA:
    a,b=10,20
    def display():
        print("I am from classA")
class ClassB(ClassA):
    c,d=12,13
    def show():
        print("I am from ClassB")
obj=ClassB
print(obj.b)
obj.show()
obj.display()

    
````
